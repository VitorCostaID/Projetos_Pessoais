## Shopping Análises

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("shopping_trends.csv")

df
```

```python
df = df.drop(columns="Customer ID")
```

```python
df.info()
```

```python
valores = df['Age'].unique()
np.sort(valores)
```

```python
bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 150]
labels = ['18-25', '25-30', '30-35', '35-40', '40-45', '45-50', '50-55', '55-60', '60-65', '65-70', '70+']

df['faixas'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
contagem = df['faixas'].value_counts()
print(contagem)
```

```python
contagem_percentual = round(df['faixas'].value_counts(normalize=True) * 100, 2)
print(contagem_percentual)
```
