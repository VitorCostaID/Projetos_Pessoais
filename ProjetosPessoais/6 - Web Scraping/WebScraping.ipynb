## Web Scraping 

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://pt.wikipedia.org/wiki/Lista_de_pa%C3%ADses_por_PIB_(Paridade_do_Poder_de_Compra)"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
soup.find_all('th')
```

```python
world_titles = soup.find_all('thead')[0]
print(world_titles)
```

```python
world_table_titles = [title.text.strip() for title in world_titles]

world_table_titles
```
